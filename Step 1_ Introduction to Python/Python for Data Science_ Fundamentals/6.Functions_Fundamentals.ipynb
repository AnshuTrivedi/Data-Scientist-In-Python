## 6.Extract Values From Any Column

As a reminder, our data set is structured as a list of lists. To generate a frequency table for a certain column, we could:

Extract the values of the column in a separate list.
Generate a frequency table for the elements of that list.

One thing we can try is to create a separate function for each of these two tasks:

* A function that extracts the values for any column we want in a separate list; and
* A function that generates a frequency table for a list.

```python
opened_file = open('C:/Users/krishna/Desktop/JUPYTE~1/Data quest/Step1I~1/PYTHON~1/AppleStore.csv',encoding='utf-8')
from csv import reader
read_file = reader(opened_file)
apps_data = list(read_file)

def extract(index):                               # def to extract column values of one column
    column=[]
    for row in apps_data[1:]:
        value=row[index]
        column.append(value)
        
    return column
    
geners=extract(11) 
print(geners)
```

```python
def freq_table(content):
    freq_table={}
    for i in content:
        if i in freq_table:
            freq_table[i] +=1
        else:
            freq_table[i]=1
    return freq_table
genres_ft=freq_table(geners)
print(genres_ft)
```

## 8.Writing a Single Function

* Write a function named freq_table() that generates a frequency table for any column in our iOS apps data set.
 * single function to extract column and count its items

```python
def freq_table(index):
    column_list=[]
    freq_items={}                                     #single function to extract column and count its items
    for row in apps_data[1:]:
        v=row[index]
        column_list.append(v)
        if v in freq_items:
            freq_items[v]+=1
        else:
            freq_items[v]=1
    return freq_items

freq_table(11)                              # 11th column
```

## 9.Reusability and Multiple Parameters

**Recall from the previous exercise that we iterated directly over the apps_data data set inside the function's definition:**

*  Iterating over apps_data inside the function's definition constrains us to use freq_table() only for the apps_data data set. 

*  If we had another data set or wanted to save the function to use it later for another project, freq_table() would be useless because it's designed to iterate only over the apps_data variable.

### So instead of having freq_table(index), we'd need something like freq_table(index, data_set)  that is, we'd need to create the function with two parameters.

* One of the key aspects that make functions great is `reusability.` 
* Python allows us to use `multiple parameters` for the functions we create.

```python
def freq_table(index,data_set):         # two parameters dataset and index
    frequency_table = {}
    
    for row in data_set[1:]:
        value = row[index]
        if value in frequency_table:
            frequency_table[value] += 1
        else:
            frequency_table[value] = 1
            
    return frequency_table

ratings_ft=freq_table(data_set=apps_data,index=7)
print(ratings_ft)
```

## 10.Keyword and Positional Arguments

When we use the syntax `subtract(a=10, b=7)` or `subtract(b=7, a=10)`, we pass in the arguments 10 and 7 using the variable names a and b. For this reason, they are called `named arguments`, or more commonly, `keyword arguments.`

* When we use keyword arguments, the `order we use to pass in the arguments doesn't make any difference.`

* Arguments that are passed by position are called `positional arguments.`
* when we use subtract(10, 7) or subtract(7, 10), we're not explicit about what arguments correspond to what parameters. First parameter mapped to first argument by position.

```python
ratings_ft=freq_table(data_set=apps_data,index=7)  # keyword arguments

ratings_ft=freq_table(7,apps_data)                 # positional arguments
```


## 11.. Combining Functions

```python
def find_sum(a_list):
    a_sum = 0
    for element in a_list:
        a_sum += float(element)
    return a_sum

def find_length(a_list):
    length = 0
    for element in a_list:
        length += 1
    return length


def mean(data_set,index):
    l=extract(apps_data,index)
    find_sum(l)                              # used find_sum()
    find_length(l)                          # used find_length()
    m=find_sum(l)/find_length(l)
    return m
```
